```python
from pyomo.environ import *

# Insiemi
Oggetti = range(1, n)

# Costante
M = 1 + max(sum(D[i, j] for j in Oggetti if j != i) for i in Oggetti)

# Variabili
model = ConcreteModel()

model.x = Var(Oggetti, domain=Binary)
model.y = Var(Oggetti, Oggetti, domain=Binary)
model.z = Var(domain=NonNegativeReals)

# Funzione obiettivo
model.obj = Objective(expr=model.z, sense=maximize)

# Vincoli
model.maxmin = Constraint(Oggetti, rule=lambda model, i: model.z <= sum(model.y[i, j] * D[i, j] for j in Oggetti if j != i) + M * (1 - model.x[i]))

model.cardinalita = Constraint(expr=sum(model.x[i] for i in Oggetti) == m)

model.simmetria = Constraint(Oggetti, Oggetti, rule=lambda model, i, j: model.y[i, j] == model.y[j, i])

model.attivazioneI = Constraint(Oggetti, Oggetti, rule=lambda model, i, j: model.y[i, j] <= model.x[i])

model.attivazioneJ = Constraint(Oggetti, Oggetti, rule=lambda model, i, j: model.y[i, j] <= model.x[j])

model.attivazioneY = Constraint(Oggetti, Oggetti, rule=lambda model, i, j: model.x[i] + model.x[j] <= model.y[i, j] + 1)

model.taglio = Constraint(Oggetti, rule=lambda model, i: sum(model.y[i, j] for j in Oggetti if j != i) <= (m - 1) * model.x[i])

# Risoluzione
solver = SolverFactory('cplex')

solver.solve(model)

# Stampa della soluzione
print('Valore obiettivo:', model.z.value)
print('Stabilimenti aperti:', [i for i in Oggetti if model.x[i].value == 1])
print('Assegnazioni:', [(i, j) for i in Oggetti for j in Oggetti if model.y[i, j].value == 1])
```
